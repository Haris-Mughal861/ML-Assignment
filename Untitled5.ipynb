```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
matches_played=np.array([14,28,42,56,70,84,98])
runs_scored=np.array([468,691,522,523,492,444,345])
slope, intercept=np.polyfit(matches_played,runs_scored,1)
regression_line=slope*matches_played+intercept
plt.scatter(matches_played, runs_scored, label="Runs Scored vs. Matches Played", color="blue")
plt.plot(matches_played, regression_line, label="Linear Regression", color="red")
plt.xlabel("Runs Scored")
plt.ylabel("Matches Played")
plt.title("Linar Regression:Babar Azam's Runs vs. No of t20i matches Played")
plt.legend()
plt.grid(True)
plt.show()
```
